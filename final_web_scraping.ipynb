**Function: Get list of images from each category**

```python
import requests 
from bs4 import BeautifulSoup 
import urllib.request
```

```python
list_of_categories = [
"tshirt",
"shorts",
"shirt",
"mens+suit",
"dress",
"pants",
"long+sleeve+tshirt",
"blouse",
]
```

```python
list_of_gen_links = []


for item in list_of_categories:
  list_of_gen_links.append(f"https://www.myer.com.au/search?query={item}")

print(list_of_gen_links)
```

```python
def getdata(url): 
    r = requests.get(url) 
    return r.text 

def get_list_image_websites(url):
  #"""Function to obtain the links for each of the images with the specified url"""
  htmldata = getdata(url)
  soup = BeautifulSoup(htmldata, 'html.parser') 

  image_url_list = []
  for item in soup.find_all('img'):
    image_url_list.append(item['src'])
  
  return image_url_list
```

```python
# place lists of links of websites into the array
categorical_links = []

for item in list_of_gen_links:
  the_added_list = []
  list_of_items = get_list_image_websites(item)
  for innerItem in list_of_items:
    if "16x20" in innerItem or '/static' in innerItem:
      print()
      #list_of_items.remove(innerItem)
    else:

      print(innerItem)
      the_added_list.append(innerItem)

  categorical_links.append(the_added_list)

print(categorical_links)
```
